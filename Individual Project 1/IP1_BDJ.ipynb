5.	Remove all the symbols (\$\,)from the 'Price' column and convert  and update it to  64-bit int numbers using astype(int64) Hint. Look for string functions like replace/strip). The data frame will be look like after that:

```python
car_sales['Price'] = car_sales['Price'].astype('str')
car_sales['Price'] = car_sales['Price'].str.strip('$')
car_sales['Price'] = car_sales['Price'].replace(',', '', regex = True)
car_sales['Price'] = car_sales['Price'].astype('int64')
```

6.	Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column look up the cumsum() function.

```python
total_sales = car_sales['Price'].cumsum()
car_sales['Total Sales in Dollars'] = total_sales
car_sales.head()
```

7.	Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today). Check out date_range function and keep the periods=len(car_sales). The data frame will be look like that:

```python
sale_date = pd.date_range(start='2/13/2025', end='2/22/2025')
print(len(sale_date), len(car_sales))

car_sales['Sale Date'] = sale_date
car_sales.head()
```

8.	Use the matplotlib line plot function to plot the 'Price' column versus the 'Total Sales' column (Give proper title and appropriate labels.)

```python
plt.title("Total Sales over Price")
plt.xlabel("Sales")
plt.ylabel("Total Sales")
plt.plot(car_sales['Price'], car_sales['Total Sales in Dollars'])
```

12.	Create a subplot of multiple plots ( step 8,9,10,11).

```python
fig, axs = plt.subplots(2, 2)
fig.set_figheight(10)
fig.set_figwidth(10)

axs[0, 0].plot(car_sales['Price'], car_sales['Total Sales in Dollars'])
axs[0, 0].set_title("Total Sales over Price")

axs[0, 1].scatter(car_sales['Odometer (KM)'], car_sales['Price'])
axs[0, 1].set_title("Odometer in KM vs. Price")

axs[1, 0].scatter(car_sales['Make'], car_sales['Odometer (KM)'])
axs[1, 0].set_title("Car Make vs Odometer in KM")

axs[1, 1].hist(car_sales['Make'])
axs[1, 1].set_title("Car Make Histogram")
```

16.	Find out Pearson correlation between 'Total Sales','Price','Odometer (KM)'. Comment on the correlation 

```python
car_sales_p = car_sales[['Total Sales in Dollars', 'Price', 'Odometer (KM)']].copy()
sns.heatmap(car_sales_p.corr(), annot = True)
```

As seen above, most variable have a weak correlation, but as Price increases, there is a relatively strong negative correlation with the kilometers on a car (and vice versa). This makes sense, as cars depreciate over time/use, and kilometers traveled represents that.
